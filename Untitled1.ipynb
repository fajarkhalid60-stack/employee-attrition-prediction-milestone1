```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
print("Original shape:", df.shape)

df_clean = df.drop(["EmployeeCount", "StandardHours", "Over18", "EmployeeNumber"], axis=1)


df_clean["Attrition"] = df_clean["Attrition"].map({"Yes": 1, "No": 0})
df_clean["Gender"] = df_clean["Gender"].map({"Male": 1, "Female": 0})
df_clean["OverTime"] = df_clean["OverTime"].map({"Yes": 1, "No": 0})


df_encoded = pd.get_dummies(df_clean, columns=["BusinessTravel", "Department", "EducationField", "JobRole", "MaritalStatus"], drop_first=True)


X = df_encoded.drop("Attrition", axis=1)
y = df_encoded["Attrition"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)
print("Training class distribution:", y_train.value_counts().to_dict())
```
